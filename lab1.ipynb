## Sine Wave - Moku Go

<img src="lab1SineScreenshot.png" alt="Sine Wave" width="50%">
<img src="lab1_sineFFTscreenshot.png" alt="Sine FFT" width="50%">


```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import math
```

```python
# Plot Sine Wave
sine = pd.read_csv("lab1_sineTraces.csv", skiprows=7, sep=", ")
sine['Time (s)'] = sine['Time (s)'] * 10**3
sine.rename(columns={'Time (s)' : 'Time (ms)'}, inplace=True)

fig = px.line(sine, x='Time (ms)', y='Voltage (V)')
fig.show()
```

```python
# Plot Sine Wave FFT Full Data Set
sineFFT = pd.read_csv("lab1_sineFFTtraces.csv", skiprows=7, sep=", ")
sineFFT['Freq (kHz)'] = sineFFT['Freq (Hz)'] * 10**5

fig = px.line(sineFFT, x='Freq (kHz)', y='FFT (dBm)')
fig.show()
```

```python
# Plot Sine Wave FFT 20 kHz
sineFFT = pd.read_csv("lab1_sineFFTtraces.csv", skiprows=7, sep=", ")
sineFFT['Freq (kHz)'] = sineFFT['Freq (Hz)'] * 10**5
sineFFT = sineFFT[sineFFT['Freq (kHz)'] < 20]

fig = px.line(sineFFT, x='Freq (kHz)', y='FFT (dBm)')
fig.show()
```

```python
# Plot Sine Wave FFT 20 kHz as V
sineFFT['V'] = math.sqrt(50/1000) * 10**(sineFFT['FFT (dBm)'] / 20)

fig = px.line(sineFFT, x='Freq (kHz)', y='V')
fig.show()
```
